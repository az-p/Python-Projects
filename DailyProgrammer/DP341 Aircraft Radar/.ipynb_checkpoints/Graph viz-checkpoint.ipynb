```python
%matplotlib inline 
import networkx as nx
import matplotlib.pyplot as plt

urls = {'reddit': ['reddit_child_1', 'reddit_child_2'],
        'reddit_child_1': ['rc1_child_1', 'rc1_child_2']}

G = nx.DiGraph(urls)
#plt.subplot(122)
nx.draw(G, with_labels=True, font_weight='bold')

plt.show()
```
